### Part II - Key Insights: The Egyptian Used Car Market




#### Investigation Overview

This presentation summarizes the key insights derived from the analysis
of the used car market in Egypt. The primary goals were to identify
market dominance by brand and model, analyze price variations by city,
and understand price trends over time.

**Summary of Key Insights:** 
1. **Market Dominance:** `Asian manufacturers, particularly Nissan, Toyota, and Kia, dominate the market volume, with the **Nissan Sunny** being the top model.`
2. **Geographic Premium:** `Zamalek & Tagamo3 - New Cairo exhibits the highest maximum used-car prices, reflecting significant regional valuation differences.`
3. **Which transmission type is more popular?** `**Automatic transmission** is more popular than **manual transmission**`
4. **Price distribution respect to the transmission type:** ` Cars with automatic transmissions generally have much higher price variability, including several very expensive outliers that skew the distribution significantly upward.`
5. **What are the 5 frequent cars in the market?, and how their prices vary over mileage?** `The top 5 frequent cars in the market are: Sunny, Corolla, Cerato, Tipo, and Logan.` `Some models (e.g., Cerato, Corolla) show a broader range of prices when kilometers are low`



#### Dataset Overview

The analysis is based on a cleaned dataset of 8912 used car listings,
featuring key variables such as Brand, Model, Year, Kilometers Driven
(km), and Price \[3\].

#### The 5 frequent cars in the market and how their prices vary over mileage?

* The Fucet graph shows the price variance over mileage for the top 5 frequent cars in the market.

```python
display(Image(filename='price_km_models.png'))
```

*   Observation: 
    - There is a consistent negative correlation between price and kilometers driven for all five car models: as the number of kilometers increases, the price tends to decrease. _This confirms the general expectation that higher mileage leads to lower resale values for used cars.​_
    - Spread of prices for each model is different at low mileage. Some models (e.g., Cerato, Corolla) show a broader range of prices when kilometers are low, possibly due to variations in year, trim, or condition. Others like Sunny appear more clustered and predictable.
    - A few outlier points (e.g., unusually expensive cars with higher mileage) are visible in some panels. These may represent rare trim levels, recent major maintenance, errors, or over estimations.
