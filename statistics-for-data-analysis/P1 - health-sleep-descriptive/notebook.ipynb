```python
p_mean = df['Physical Activity Level'].mean()
p_median = df['Physical Activity Level'].median()
```

```python
p_mode = df['Physical Activity Level'].mode()[0]
p_mode
```

```python
Physical_stats_ = {'mean': p_mean, 'median': p_median, 'mode': p_mode}
Physical_stats = pd.DataFrame(Physical_stats_ , index=['Physical Activity Level'])
Physical_stats
```

```python
fig,x = plt.subplots(figsize=(1,3))
df['Daily Steps'].plot(kind='box')
plt.title('Daily Steps')
plt.ylabel('Steps')
plt.grid()
plt.show()
```

```python
fig,x = plt.subplots(figsize=(5,4))
df['Heart Rate'].plot(kind='hist')
plt.title('Heart Rate')
plt.xlabel('bpm')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```
